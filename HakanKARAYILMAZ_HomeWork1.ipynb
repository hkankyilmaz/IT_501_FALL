```python
#question 3

import math

#@param x1 <int> , <float> = x axis of first point
#@param y1 <int> , <float> = y axis of first point
#@param x2 <int> , <float> = x axis of second point
#@param y1 <int> , <float> = x axis of second point
#@return  <int> , <float>

def calculateDistanceBetweenTwoPoint(x1,y1,x2,y2):
    return math.sqrt( ( (x2-x1)**2 ) + ((y2-y1)**2) )

calculateDistanceBetweenTwoPoint(1,1,4,5)
```

```python
#question 5

import random

#@param l <list> = list that have numbers
#@param n <int> <float> = number
#@return list

def listConvertor(l,n):
    newList = []
    for x in l:
        if n>x:
            newList.append(x+random.randint(n-x,n+1905))
        else:
            newList.append(x)
    return newList

listConvertor([23,33,44,55,66],35)




```

```python
#question 6

#@param l <list> = list that have numbers
#@param n <int> <float> = number
#@return n <int> <float> = number

def countNumber(l,n):
    count = 0
    for item in l:
        if item==n:
            count= count+1
        else:
            pass
    return count

countNumber([33,21,99,1,33,57,32,123,55,33,70,33],33)
```

```python
#question 9

#@param l1 <list> = list 
#@param l2 <list> = list 
#@return list

def findEqualsValues(l1,l2):
    result=[]
    for i in l1:
        for j in l2:
            if (i==j and i not in result):
                result.append(i)
    return result

findEqualsValues([1,2,3,4,5,6,7],[6,2,5,9,8,6,6])
```

```python
#question 10

#@param l <list> = list 
#@return list

import random

def shuffleList(l):
    length= len(l)
    shuffledList =[]
    usedNumbers= []
    while( len(l)!=len(shuffledList) and l!=shuffledList ):
        ranNum = random.randint(0,length-1)
        if(l[ranNum] not in usedNumbers):
            shuffledList.append(l[ranNum])
            usedNumbers.append(l[ranNum])
        else:
            pass
    return shuffledList

shuffleList([1,2,3,4,5,6,7])


```
